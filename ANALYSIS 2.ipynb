```python
import pandas as pd
import numpy as np
import dask.dataframe as dd
df = pd.read_csv('data_1.2.3.csv', low_memory = False)
# df = dd.read_csv('merged_gcs_included.csv', dtype={'BIB': 'object',
#        'Today': 'object'})
```

```python
# df = df.drop_duplicates(subset=['Rider', 'Stage', 'Date'], keep='last') # fixes weird issue with duplicate entries for some races
pd.set_option('display.max_columns', None)
df[(df['Rider'] == 'FROOME Chris') & (df['Year'] < 2019)]['Sprint_pts'].to_numpy().sum()
```

```python
#df['Sprint_current_season'] = 0
#for rider in df.Rider.unique():
#    df['Sprint_current_season'][i] = df[(df['Rider'] == df['Rider'][i]) & (df['Year'] < df['Year'][i])]['Sprint_pts'].to_numpy().sum()
big_df = pd.DataFrame()
subset = []
i = 1
for rider in df.Rider.unique():
    subset = df[df['Rider'] == rider].sort_values("Date")
    print('Rider number '+str(i)+' out of '+str(len(df.Rider.unique()))+' starting now. His name is '+str(rider)+' .')
    for year in df.Year.unique():
        col_name1 = 'Sprint_{}_sum'
        subset[col_name1.format(year)] = subset[subset['Year'] == year]['Sprint_pts'].to_numpy().sum()

        col_name2 = 'Climb_{}_sum'
        subset[col_name2.format(year)] = subset[subset['Year'] == year]['Climb_pts'].to_numpy().sum()

        col_name3 = 'Oneday_{}_sum'
        subset[col_name3.format(year)] = subset[subset['Year'] == year]['Oneday_pts'].to_numpy().sum()

        col_name4 = 'GC_{}_sum'
        subset[col_name4.format(year)] = subset[subset['Year'] == year]['GC_pts'].to_numpy().sum()


    big_df = big_df.append( subset)
    i = i + 1
    subset = []
```
