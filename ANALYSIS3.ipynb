```python
def sum_until_this_year(this_year, discipline):
    df[str(discipline)+'_sum_until_'+str(this_year)] = 0
    for index in range(2011, this_year):
        df[str(discipline)+'_sum_until_'+str(this_year)] = df[str(discipline)+'_'+str(index)+'_sum']
```

```python
for year in test.Year.unique():
    df = pd.read_csv("data_1.4.3_"+str(year)+".csv", low_memory = False)
    df["Sprint_current_season"] = 0
    for disciplina in ["Sprint", "Climb", "Oneday", "GC"]:
        df[str(disciplina)+'_sum_until_'+str(year)] = sum_until_this_year(year, disciplina)
        df[str(disciplina)+'_current_season'] = df[str(disciplina)+'_cumsum'] - df[str(disciplina)+'_sum_until_'+str(year)]
    print('Year '+str(year)+' finished.')
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv('data_1.4.3_2020.csv', low_memory = False)
# df = dd.read_csv('merged_gcs_included.csv', dtype={'BIB': 'object',
#        'Today': 'object'})
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
df[(df['Rnk'] == 1)].head(20).sort_values("Date")
```

```python
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
disciplines = ['Sprint', 'Climb', 'Oneday', 'GC']
for year in years:
    df = pd.read_csv('data_1.4.3_'+str(year)+'.csv', low_memory = False)
    for entry in disciplines:
        cols = []
        cols = [col for col in df if col.startswith(str(entry)+'_2')]
        keep = year - 2011
        # cols.remove(str(entry)+'_'+str(year)+'_sum')
        cols = cols[:keep]
        df[str(entry)+'_current_season'] = df[str(entry)+'_cumsum']
        for col in cols:
            df[str(entry)+'_current_season'] = df[str(entry)+'_current_season'] - df[col]
    df.to_csv('data_1.5_'+str(year)+'.csv')
    print("Year"+str(year)+" finished.")
```
