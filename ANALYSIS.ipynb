```python
import pandas as pd
import numpy as np
import dask.dataframe as dd
df = pd.read_csv('merged_gcs_included.csv', low_memory = False)
# df = dd.read_csv('merged_gcs_included.csv', dtype={'BIB': 'object',
#        'Today': 'object'})
```

```python
# setting up columns
pd.set_option('display.max_columns', None)
df['Sprint_pts'] = 0
df['Climb_pts'] = 0
df['Oneday_pts'] = 0
df['GC_pts'] = 0
# CODING GC AS -1!!!!
df['Stage'] = df['Stage'].replace(['GC'],-1)
df.Stage = df.Stage.astype(int)
```

```python
for i in range(len(df)):
    # 1. flat stage, 2. not a gc result
    if df.ProfileScore[i] < 50 and df.Stage[i] > 0:
        df['Sprint_pts'][i] = df['Pnt'][i]
    # 1. mountain stage, 2. not a gc result
    elif df.ProfileScore[i] >= 50 and df.Stage[i] > 0:
        df['Climb_pts'][i] = df['Pnt'][i]
    elif (df.Stage[i] == 0):
        df['Oneday_pts'][i] = df['Pnt'][i]
    elif (df.Stage[i] == -1):
        df['GC_pts'][i] = df['Pnt'][i]
```

```python
import datetime
# adding cols for year, month, day
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['Month'] = pd.DatetimeIndex(df['Date']).month
df['Day'] = pd.DatetimeIndex(df['Date']).day
df.to_csv('data_1.2.3.csv')
df
```

```python
# computing the sums in the current season
subset = []
subset2 = []
ddf2 = pd.DataFrame(columns=df.columns) 
app_years = pd.DataFrame(columns=df.columns) 

for rider in df.Rider.unique():
    subset2 = df[df['Rider'] == rider].sort_values("Date")

    for year in subset2.Year.unique():
        # compute cumsums for the the current season
        subset = subset2[subset2['Year'] == year].sort_values("Date") # take only current year

        subset['Sprint_current_season'] = subset['Sprint_pts'].cumsum()
        subset['Climb_current_season'] = subset['Climb_pts'].cumsum()
        subset['Oneday_current_season'] = subset['Oneday_pts'].cumsum()
        subset['GC_current_season'] = subset['GC_pts'].cumsum()

        subset['Sprint_'+str(year)+'_season'] = subset['Sprint_pts'].cumsum()
        subset['Climb_'+str(year)+'_season'] = subset['Climb_pts'].cumsum()
        subset['Oneday_'+str(year)+'_season'] = subset['Oneday_pts'].cumsum()
        subset['GC_'+str(year)+'_season'] = subset['GC_pts'].cumsum()

        app_years = app_years.append( subset)
        subset = []


    ddf2 = ddf2.append( app_years)
    
    
```
